```python
# importing necessary libs
import pandas as pd #data manipulation
import numpy as np #numerical ppython -linear alg cals - matrices - vectors
import matplotlib.pyplot as plt #visualization
```

```python
#feature and target selection
X =df[['Temperature (°C)']]
Y = df['Energy Consumption (kWh)']
```

```python
X.shape
```

```python
Y.shape
```

```python
# import ML 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#split it to trainn test
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.8,random_state=42)#randomm_state is an arrgument
#random_state is for getting same reproducibility
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
#instantiate the model
model = LinearRegression()
```

```python
#train the model
model.fit(X_train,Y_train)
```

```python
#prediction
y_pred = model.predict(X_test)
print(f'The model predictions are \n{y_pred}')
```

```python
len(y_pred)
```

```python
#evaluate the model
mse = mean_squared_error(Y_test,y_pred)
mse
```

```python
r2 = r2_score(Y_test,y_pred)
r2
```

```python
plt.scatter(X_test, Y_test)
plt.scatter(X_test, y_pred)
```

```python
plt.scatter(X_test, Y_test, label='Actual Data')
plt.scatter(X_test, y_pred, label='Predicted Data')
plt.legend()
```

```python
# to save the model
import joblib
joblib.dump(model,'HKR_MODEL.pkl')#pkl-pickel file formate #
```
